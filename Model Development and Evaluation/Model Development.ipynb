```python
import pandas as pd 
import numpy as np 
import sklearn 
import matplotlib.pyplot as plt 
import seaborn as sns
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
encoder = LabelEncoder()

# Reading the individual Samples.
sample1 = pd.read_csv('sample1.csv')
sample2 = pd.read_csv('sample2.csv')
sample3 = pd.read_csv('sample3.csv')
sample4 = pd.read_csv('sample4.csv')
sample5 = pd.read_csv('sample5.csv')
sample6 = pd.read_csv('sample6.csv')
sample7 = pd.read_csv('sample7.csv')
sample8 = pd.read_csv('sample8.csv')
sample9 = pd.read_csv('sample9.csv')
sample10 = pd.read_csv('sample10.csv')
sample11 = pd.read_csv('sample11.csv')
sample12 = pd.read_csv('sample12.csv')

# dropping the unnecessary columns
sample1.drop('Unnamed: 0',axis =1 , inplace = True)
sample2.drop('Unnamed: 0',axis =1 , inplace = True)
sample3.drop('Unnamed: 0',axis =1 , inplace = True)
sample4.drop('Unnamed: 0',axis =1 , inplace = True)
sample5.drop('Unnamed: 0',axis =1 , inplace = True)
sample6.drop('Unnamed: 0',axis =1 , inplace = True)
sample7.drop('Unnamed: 0',axis =1 , inplace = True)
sample8.drop('Unnamed: 0',axis =1 , inplace = True)
sample9.drop('Unnamed: 0',axis =1 , inplace = True)
sample10.drop('Unnamed: 0',axis =1 , inplace = True)
sample11.drop('Unnamed: 0',axis =1 , inplace = True)
sample12.drop('Unnamed: 0',axis =1 , inplace = True)

# Dividing the sample into training and testing samples. 
x1,y1 = sample1.iloc[:,:-1],sample1.iloc[:,-1]
x2,y2 = sample2.iloc[:,:-1],sample2.iloc[:,-1]
x3,y3 = sample3.iloc[:,:-1],sample3.iloc[:,-1]
x4,y4 = sample4.iloc[:,:-1],sample4.iloc[:,-1]
x5,y5 = sample5.iloc[:,:-1],sample5.iloc[:,-1]
x6,y6 = sample6.iloc[:,:-1],sample6.iloc[:,-1]
x7,y7 = sample7.iloc[:,:-1],sample7.iloc[:,-1]
x8,y8 = sample8.iloc[:,:-1],sample8.iloc[:,-1]
x9,y9 = sample9.iloc[:,:-1],sample9.iloc[:,-1]
x10,y10 = sample10.iloc[:,:-1],sample10.iloc[:,-1]
x11,y11= sample11.iloc[:,:-1],sample11.iloc[:,-1]
x12,y12 = sample12.iloc[:,:-1],sample12.iloc[:,-1]
# Encoding the output Variables using Label Encoder
y1 = encoder.fit_transform(y1)
y2 = encoder.fit_transform(y2)
y3 = encoder.fit_transform(y3)
y4 = encoder.fit_transform(y4)
y5 = encoder.fit_transform(y5)
y6 = encoder.fit_transform(y6)
y7 = encoder.fit_transform(y7)
y8 = encoder.fit_transform(y8)
y9 = encoder.fit_transform(y9)
y10 = encoder.fit_transform(y10)
y11 = encoder.fit_transform(y11)
y12 = encoder.fit_transform(y12)

# Creatring 12 instances of SVM to create 12 base estimators.
svm1 = SVC()
svm2 = SVC()
svm3 = SVC()
svm4 = SVC()
svm5 = SVC()
svm6 = SVC()
svm7 = SVC()
svm8 = SVC()
svm9 = SVC()
svm10 = SVC()
svm11 = SVC()
svm12 = SVC()

# Fitting the models on different samples
svm1.fit(x1,y1)
svm2.fit(x2,y2)
svm3.fit(x3,y3)
svm4.fit(x4,y4)
svm5.fit(x5,y5)
svm6.fit(x6,y6)
svm7.fit(x7,y7)
svm8.fit(x8,y8)
svm9.fit(x9,y9)
svm10.fit(x10,y10)
svm11.fit(x11,y11)
svm12.fit(x12,y12)

#Creating test data for the models: 
# using 200 randomly sampled records from sample 12 as the test data for the first 11 models
# and using 200 randomly sampled records from sample 11 as the testing set for the 12th model.
test_index = []
for i in range(0,200):
    ind = np.random.randint(0,len(x12))
    test_index.append(ind)
x_test = x12.loc[test_index]
y_test = y12[test_index]
x_test1 = x11.loc[test_index]
y_test1 = y11[test_index]
# finding the accuracy score of the models on test data.
models = [svm1,svm2,svm3,svm4,svm5,svm6,svm7,svm8,svm9,svm10,svm11,svm12]
accuracy = []
for i,model in enumerate(models):
    if(i <= 11):
        acc = model.score(x_test,y_test)
    else:
        acc = model.score(x_test1,y_test1)
    print("Accuracy of model {} is : {}".format(i+1,acc))
    accuracy.append(acc)

# pickling the modles.
pickle.dump(svm2,open('svm2.sav','wb'))
pickle.dump(svm3,open('svm3.sav','wb'))
pickle.dump(svm4,open('svm4.sav','wb'))
pickle.dump(svm5,open('svm5.sav','wb'))
pickle.dump(svm6,open('svm6.sav','wb'))
pickle.dump(svm7,open('svm7.sav','wb'))
pickle.dump(svm8,open('svm8.sav','wb'))
pickle.dump(svm9,open('svm9.sav','wb'))
pickle.dump(svm10,open('svm10.sav','wb'))
pickle.dump(svm11,open('svm11.sav','wb'))
pickle.dump(svm12,open('svm12.sav','wb'))
```
